## Задание 1
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

**Нормальное распределение**

mu - координата экстремума по оси Х, sigma - отклонение по оси Y.

```python
x = np.linspace(0,4,100)
for mu in [1.0, 2.0]:
    for sigma in [0.1,0.3,0.5]:
        norm_rv = sts.norm(loc=mu, scale=sigma)
        norm_rv.rvs(size=10)
        pdf = norm_rv.pdf(x)
        plt.plot(x, pdf)
        plt.ylabel('$f(x)$')
        plt.xlabel('$x$')
        plt.title(f'mu={mu}, sigma={sigma}')
        plt.show()
```

**Экспоненциальное распределение**
lam - смещение от начала координат

```python
x = np.linspace(0,5,1000)
for lam in [0.1, 0.5, 1, 1.5, 2]:
    expon_rv = sts.expon(lam)
    pdf = expon_rv.pdf(x)
    plt.plot(x, pdf)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.title(f'lam={lam}')
    plt.show()
```

**Распределение Стьюдента**

на распределение Стьюдента влияет только диапазон Х. По графику ниже видно, что n - всего лишь влияет на точность построения.

```python
x = np.linspace(-5,5,100)
t_rv = sts.t(10)
t_rv.rvs(10)
pdf = t_rv.pdf(x)
plt.plot(x, pdf)
t_rv = sts.t(20)
t_rv.rvs(20)
pdf = t_rv.pdf(x)
plt.plot(x, pdf)
t_rv = sts.t(6000)
t_rv.rvs(6000)
pdf = t_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

plt.show()
```
